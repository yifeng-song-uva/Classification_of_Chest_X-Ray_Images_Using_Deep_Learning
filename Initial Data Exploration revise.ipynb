```python
import os
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

data_path = r'D:\\machine learning\\final project\\rscbjbr9sj-2\\ChestXRay2017\\chest_xray'
save_path = r"D:\machine learning\final project\rscbjbr9sj-2\ChestXRay2017\figure and table"

if not os.path.exists(save_path):
    os.makedirs(save_path)
```

#### Load the data

```python
image_data = []
for dataset in ['train', 'test']:
    for class_name in ['NORMAL', 'PNEUMONIA']:
        folder = os.path.join(data_path, dataset, class_name)
        for img_file in os.listdir(folder):
            if img_file.lower().endswith('.jpeg'):
                file_lower = img_file.lower()
                if class_name == 'NORMAL':
                    sublabel = 'NORMAL'
                elif 'bacteria' in file_lower:
                    sublabel = 'BACTERIA'
                elif 'virus' in file_lower:
                    sublabel = 'VIRUS'
                else:
                    sublabel = None
                image_data.append({
                    'image_path': os.path.join(folder, img_file),
                    'label': class_name,
                    'sublabel': sublabel,
                    'dataset': dataset
                })

df = pd.DataFrame(image_data)
sublabel_categories = ['NORMAL', 'BACTERIA', 'VIRUS']
df['sublabel'] = pd.Categorical(df['sublabel'], categories=sublabel_categories, ordered=True)
print(df['sublabel'].value_counts(dropna=False))
print(df[['label', 'sublabel']].drop_duplicates())
```

#### Summary statistics

```python
summary_1 = pd.DataFrame.from_dict(Counter(df.dataset), orient="index")
summary_1.columns = ['Count']
summary_1.loc[:,'Percentage'] = summary_1.Count / df.shape[0] * 100
summary_1 = summary_1.round(2)
summary_1.loc[:,'Percentage'] = summary_1.loc[:,'Percentage'].astype(str) + "%"
# summary_1 = summary_1.reset_index().rename(columns={'index': 'dataset'})
summary_1.index.name = "dataset"
summary_1
```

```python
part2 = df.groupby(['dataset', 'label']).agg({'image_path': 'count'}) 
summary_2 = part2.join(summary_1.loc[:,['Count']], on=['dataset'], how='left')
summary_2.loc[:,'Percentage'] = summary_2.image_path / summary_2.Count * 100
summary_2 = summary_2.round(2)
summary_2 = summary_2.drop(['Count'], axis=1).rename(columns={'image_path':'Count'})
summary_2.loc[:,'Percentage'] = summary_2.loc[:,'Percentage'].astype(str) + "%"
summary_2
```

```python
part3 = df.groupby(['dataset', 'sublabel'], observed=False).agg({'image_path': 'count'}) 
summary_3 = part3.join(summary_1.loc[:,['Count']], on=['dataset'], how='left')
summary_3.loc[:,'Percentage'] = summary_3.image_path / summary_3.Count * 100
summary_3 = summary_3.round(2)
summary_3 = summary_3.drop(['Count'], axis=1).rename(columns={'image_path':'Count'})
summary_3.loc[:,'Percentage'] = summary_3.loc[:,'Percentage'].astype(str) + "%"
summary_3
```

#### Image Size Statistics

```python
def clean_summary_table(table):
    table2 = table[table.index != 'count'].copy()
    table2 = table2.round({'X': 0, 'Y': 0, '#_of_pixels': 0, 'aspect_ratio': 2})
    for cn in ['X', 'Y', '#_of_pixels']:
        table2.loc[:, cn] = table2.loc[:, cn].astype(int).astype(str)
    return table2

img_size_list = []
for i in range(df.shape[0]):
    img_path = df.iloc[i]['image_path']
    img = Image.open(img_path)
    img_size_list.append(img.size)

img_size = pd.DataFrame(img_size_list, columns=['X', 'Y'])
img_size.loc[:, '#_of_pixels'] = img_size.X * img_size.Y
img_size.loc[:, 'aspect_ratio'] = img_size.X / img_size.Y
img_size.loc[:, 'dataset'] = df['dataset'].values
img_size.loc[:, 'sublabel'] = df['sublabel'].values
all_tables = [] 
for v1 in ['train', 'test']:
    for v2 in sublabel_categories:
        subset = img_size[(img_size['dataset'] == v1) & (img_size['sublabel'] == v2)]
        if len(subset) > 0:
            tbl = clean_summary_table(subset[['X', 'Y', '#_of_pixels', 'aspect_ratio']].describe())
            print(f"{v1.upper()} | {v2} (n={len(subset)})")
            print(tbl.to_string())
            print()
            tbl_to_save = tbl.copy()
            tbl_to_save.insert(0, 'Sublabel', v2)
            tbl_to_save.insert(0, 'Dataset', v1.upper())
            all_tables.append(tbl_to_save)
if all_tables:
    final_df = pd.concat(all_tables)
    final_df.to_csv(os.path.join(save_path, "summary_report.csv"), encoding='utf-8-sig')
```

#### Patient-Level Data Leakage Analysis

```python
import re

def extractPatientID(filename, class_name):
    if class_name == 'NORMAL':
        return re.sub(r'-\d{4}.*', '', os.path.basename(filename))
    else:
        m = re.match(r'(person\d+)', os.path.basename(filename))
        return m.group(1) if m else os.path.basename(filename)

df['patientid'] = df.apply(
    lambda row: extractPatientID(row['image_path'], row['label']), axis=1
)

trainPatients = set(df[df['dataset'] == 'train']['patientid'])
testPatients  = set(df[df['dataset'] == 'test']['patientid'])
overlapPatients = trainPatients & testPatients
leakageRate = round(len(overlapPatients) / len(testPatients) * 100, 1)

leakageSummary = pd.DataFrame({
    'Metric': ['Train patients', 'Test patients', 'Overlapping patients', 'Leakage rate (%)'],
    'Value':  [len(trainPatients), len(testPatients), len(overlapPatients), str(leakageRate) + '%']
})
print(leakageSummary.to_string(index=False))
leakageSummary.to_csv(os.path.join(save_path, "data_leakage_summary.csv"), index=False, encoding='utf-8-sig')
```

#### Class Imbalance & Weighted Cross-Entropy

```python
trainCounts = df[df['dataset'] == 'train']['sublabel'].value_counts()
N = trainCounts.sum()
K = 3
classWeights = {cls: round(N / (K * trainCounts[cls]), 4) for cls in sublabel_categories}

weightTable = pd.DataFrame({
    'Class':        sublabel_categories,
    'Train Count':  [trainCounts[c] for c in sublabel_categories],
    'Proportion (%)': [round(trainCounts[c] / N * 100, 2) for c in sublabel_categories],
    'Class Weight': [classWeights[c] for c in sublabel_categories]
})
print(weightTable.to_string(index=False))
weightTable.to_csv(os.path.join(save_path, "weightTable.csv"), index=False, encoding='utf-8-sig')
```
